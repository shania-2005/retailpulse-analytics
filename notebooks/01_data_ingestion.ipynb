# RetailPulse - Data Ingestion

## Objective

The objective of this notebook is to load the Online Retail II dataset,
merge multiple sheets into a single dataset, perform a basic inspection,
and save the merged dataset for further preprocessing.

Dataset:
Online Retail II (UCI Machine Learning Repository)

```python
# Import the required library
import pandas as pd
```

```python
# Path of the raw dataset
# (Keep the Excel file inside data/raw folder)

file_path = "../data/raw/online_retail_II.xlsx"
```

```python
# Load the first sheet (Year 2009-2010)

df_2009 = pd.read_excel(
    file_path,
    sheet_name="Year 2009-2010"
)

# Load the second sheet (Year 2010-2011)

df_2010 = pd.read_excel(
    file_path,
    sheet_name="Year 2010-2011"
)
```

```python
# Display number of rows and columns in both sheets

print("2009-2010 Shape :", df_2009.shape)
print("2010-2011 Shape :", df_2010.shape)
```

```python
# Merge both sheets into a single dataframe

merged_df = pd.concat(
    [df_2009, df_2010],
    ignore_index=True
)
```

```python
# Display first five rows

merged_df.head()
```

```python
# Display last five rows

merged_df.tail()
```

```python
# Display dataset dimensions

print("Merged Dataset Shape :", merged_df.shape)
```

```python
# Display column names

merged_df.columns
```

```python
# Display data types and missing values

merged_df.info()
```

```python
# Generate descriptive statistics for numerical columns

merged_df.describe()
```

```python
# Generate descriptive statistics for categorical columns

merged_df.describe(include="object")
```

```python
# Count missing values in each column

merged_df.isnull().sum()
```

```python
# Save merged dataset for the next step (Data Cleaning)

merged_df.to_csv(
    "../data/interim/merged_data.csv",
    index=False
)

print("Merged dataset saved successfully.")
```
