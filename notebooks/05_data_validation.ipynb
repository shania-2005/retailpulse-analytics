## Objective

The objective of this notebook is to validate the feature-engineered dataset by checking its quality, completeness, consistency, and correctness before it is used for machine learning model development and dashboard integration.

```python
import os
import sys

project_root = os.path.abspath("..")
sys.path.insert(0, project_root)
```

```python
import pandas as pd
import numpy as np


from src.data.validate_data import validate_data
```

```python
df = pd.read_csv(
    "../data/processed/feature_engineered_data.csv",
    parse_dates=["InvoiceDate"]
)

df.head()
```

```python
validation_report = validate_data(df)

validation_report
```

```python
feature_columns = [
    "Revenue",
    "Year",
    "Month",
    "MonthName",
    "Day",
    "DayOfWeek",
    "Hour",
    "Quarter",
    "IsWeekend",
    "BasketSize",
    "OrderValue"
]

print(df[feature_columns].head())
```

```python
validation_summary = pd.DataFrame({
    "Validation Check": [
        "Missing Values",
        "Duplicate Rows",
        "Negative Quantity",
        "Negative Price",
        "Negative Revenue"
    ],
    
    "Expected": [
        0,
        0,
        0,
        0,
        0
    ],
    
    "Actual": [
        df.isnull().sum().sum(),
        df.duplicated().sum(),
        (df["Quantity"] <= 0).sum(),
        (df["Price"] <= 0).sum(),
        (df["Revenue"] <= 0).sum()
    ]
})

# Create Status column
validation_summary["Status"] = np.where(
    validation_summary["Expected"] == validation_summary["Actual"],
    "Pass",
    "Fail"
)

validation_summary
```
