```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "../data/interim/cleaned_data.csv",
    parse_dates=["InvoiceDate"]
)

df.head()
```

```python
print(df.shape)

df.info()
```

### Revenue

This feature will be used for sales analysis, customer value estimation, forecasting, and inventory optimization.

```python
df["Revenue"] = df["Quantity"] * df["Price"]

df[["Quantity","Price","Revenue"]].head()
```

### Time-based Features

Time-related features are extracted from the InvoiceDate column to analyze purchasing behavior over different time periods.

```python
df["Year"] = df["InvoiceDate"].dt.year

df["Month"] = df["InvoiceDate"].dt.month

df["MonthName"] = df["InvoiceDate"].dt.month_name()

df["Day"] = df["InvoiceDate"].dt.day

df["DayOfWeek"] = df["InvoiceDate"].dt.day_name()

df["Hour"] = df["InvoiceDate"].dt.hour

df["Quarter"] = df["InvoiceDate"].dt.quarter
```

### Weekend Indicator Feature

```python
df["IsWeekend"] = (
    df["DayOfWeek"]
    .isin(["Saturday", "Sunday"])
    .astype(int)
)

df[["DayOfWeek","IsWeekend"]].head()
```

### Basket Size

Basket Size represents the total number of items purchased within a single invoice.

```python
basket_size = (
    df.groupby("Invoice")["Quantity"]
    .sum()
)

df["BasketSize"] = df["Invoice"].map(basket_size)

df[["Invoice","BasketSize"]].head()
```

### Order Value

```python
order_value = (
    df.groupby("Invoice")["Revenue"]
    .sum()
)

df["OrderValue"] = df["Invoice"].map(order_value)

df[["Invoice","OrderValue"]].head()
```
